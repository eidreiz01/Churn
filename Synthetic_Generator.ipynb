```python
# !pip install sdv
import pandas as pd
from sdv.tabular import GaussianCopula
```

```python
df = pd.read_csv("Dataset1cs add col.csv")
df.head()
```

```python
df.shape
```

```python
model = GaussianCopula()
model.fit(df)
```

```python
sample = model.sample(98000)
sample.head()
```

```python
sample.shape
```

```python
# index of 80% data with churn equal to low
to_drop = sample[sample["Churn"] == "low"].sample(frac=0.8).index
```

```python
sample.drop(to_drop, inplace=True) # dropped
```

```python
100000 - len(pd.concat([df, sample])) # need 19075 more data
```

```python
syn_data = pd.concat([df, sample, model.sample(19075)]).reset_index(drop=True)
syn_data.head()
```

```python
syn_data.shape
```

```python
syn_data.to_csv("Synthetic_Churn.csv", index=False)
```
